# 2. 재무제표 핵심
: 크게 **손익계산서 / 재무상태표 / 현금흐름표 / 자본변동표** 4가지로 구성됨

![재무]()

*   재무상태표 : 자본, 부채 등 기업의 재산 현황
*   손익계산서 : 이익과 관련된 내용
*   현금흐름표 : 기업이 사용한 현금의 움직임을 보여줌

![pbr]()

* **ROE(Return on Equity, 투자수익률)**: 자본을 가지고 *얼마나 이익을 만들었는지* 계산 -> *ROE = 이익/자본*
* **PBR(Price Book Ratio, 주가순자산비율)** : 주가가 *자본에 비해* 얼마나 높은가 -> *PBR = 주가/자본*
* **PER(Price Earning Ratio, 주가수익비율)** : *이익에 비해* 얼마나 높은가 -> *PER = 주가/이익*

# 3. 재무상태표(Balance Sheets)

![재무상대]()

* 현재의 재무상태를 보여주는 보고서
* 유동자산(Current Asset): 당장(=1년) 쓸 수 있는 돈 (1년 이내에 사용 가능한 돈)
* 비유동자산(Non-current Asset) : 묶여있는 돈
* 부채도 유동부채(Current Liability)와 비유동부채(Non-Current Liability)로 나뉨
* 회사가 당장 쓸 수 있는 돈은 얼마나 많이 확보했는지, 당장 갚아야 할 돈은 얼마인지를 재무상태표에서 읽어내야 함 → *유동비율(Current ratio)*로 표기



1.   **자산(Asset)** : 총자산=자산총액=자산총계=Total Asset
2.   **부채(Liability)** : 총부채=부채총액=부채총계=Total Liabilities
3.   **자본(Equity)** : 총자본=자본총액=자본총계=순자산=자기자본=징부가=Shareholder’s Equity



# 4. 현금상대표
: *영업현금흐름 / 투자현금흐름 / 재무현금흐름* 이렇게 3가지로 나뉨

![현금상대표]()


* **영업현금흐름(Cash Flow from Operating Activities, Operating Cash Flow, CFO)**
  : 장사를 잘 할수록, 돈을 빨리 받을수록 늘어나는 것이니 클수록 좋음

* **투자현금흐름(Cash Flow from Investing Activities, Investing Cash Flow, CFI)**
    : 공장을 짓는 시설 투자 같은 것에 들어가는 현금 → 기업이 미래를 위해 얼마나 준비하고 있는지 나타냄

* 미래를 위해 투자를 많이 할수록 마이너스가 된다.
* **재무현금흐름(Cash Flow from Financing Activities, Financing Cash Flow, CFF)**
    : 채무 상환, 배당금 지급 같은 이유로 돈이 나가면 **마이너스**, 대출을 더 많아 돈이 유입되면 **플러스**가 됨
